#### Duplicate Values

````python
# Dataset Duplicate Value Count
def duplicate_val_cnt(df):
  try:
      return print('Count of Duplicate Value is {}'.format(df.duplicated().sum()))
  except FileNotFoundError:
      print("Error: The file was not found. Please check the file path and try again.")
  except Exception as e:
      print("An error occurred:", {e})
duplicate_val_cnt(hotel_df)
````

#### Missing Values/Null Values

````python
# function to find Missing Values/Null Values Count
def missing_val_cnt(df):
  try:
      null_val = df.isnull().sum()
      # reset the index and rename the column
      after_reset_col = null_val.sort_values(ascending=False).reset_index().rename(columns={'index':'Column',0:'Null Values'})
      return after_reset_col
  except Exception as e:
      print("An error occurred:", {e})

# get Missing Values/Null Values count
missing_val_cnt(hotel_df)
````

````python
# Visualizing the missing values
def missing_val_visual(df):
  try:
      plt.figure(figsize=(10,5))
      sns.heatmap(df.isnull(),cbar=False)
      plt.title('Missing Values Heatmap')
      plt.xlabel("Column Labels")
      plt.ylabel("Row Index")
      plt.show()
  except Exception as e:
      print("An error occurred:", {e})
missing_val_visual(hotel_df)
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# make a copy of dataframe to work on df
df_copy = hotel_df.copy()
````

````python
df_copy
````

````python
# Deleting Duplicates values from dataset
df_copy.drop_duplicates(inplace=True)

# Getting count of rows and columns after removed duplicates
print(f'the number of rows: {df_copy.shape[0]} and columns: {df_copy.shape[1]} after removed duplicates')
````

````python
# Calculate the percentage of null values of each column

# function to get percentage of null values
def cal_null_val_percentage(df):
    try:
        # Calculate the percentage of null values
        percent = df.isna().sum()/df.shape[0]*100     # df.isna() or df.isnull()

        # sorting null values in descending order and reset the index and rename the column
        perc_srt = percent.sort_values(ascending=False).reset_index().rename(columns={'index':'Columns',0:'Null %'})

        return perc_srt
    except Exception as e:
        print("An error occurred:", {e})
cal_null_val_percentage(df_copy)
````

````python
# In above result we can see that column company has approx. 94% of missing value
# which is very high so droping the column company
df_copy.drop(columns=['company'],inplace=True)
````

````python
# Now check for the Other columns they have less % of null values so replace them with 0

df_copy.fillna({'agent':0,'country':'Others','children':0},inplace=True)
````

````python
# checking again for any null values
missing_val_cnt(df_copy)
````

````python
df_copy.shape
````

# **Conclusion**

1. **Total Bookings by Hotel Type**

  City hotels have a higher number of total bookings compared to resort hotels, indicating a greater preference or higher demand for city hotels.

2. **Booking Cancellation Rate**

  City hotels experience a higher cancellation rate compared to resort hotels. This suggests that guests might be more likely to cancel bookings for city hotels, potentially due to various factors like flexibility in travel plans or competitive alternatives.

3. **Average Daily Rate (ADR)**

  City hotels have a higher average daily rate compared to resort hotels. This indicates that city hotels are able to charge more per room on average, possibly due to their location and higher demand.

4. **Revenue by Hotel Type**

  City hotels generate more revenue compared to resort hotels. This is likely a combination of higher booking volumes and higher ADRs.

5. **Booking Changes by Customer Type**

  Transient customers account for the majority of booking changes, followed by transient-party customers. This suggests that transient customers are more prone to modify their bookings, which could impact operational planning and resource allocation.

6. **Lead Time by Market Segment**

  Group bookings have the longest lead time, while bookings from undefined sources have the shortest lead time. This indicates that group bookings are planned well in advance, whereas undefined bookings are made closer to the arrival date.

7. **Total Guests by Category**

  The majority of guests are adults, with children and babies making up a smaller percentage of the total guests. This demographic information is crucial for tailoring services and amenities to the predominant guest profile.

8. **Guests by Country**

  The map shows a global distribution of guests, with higher concentrations in North America, Europe, and parts of Asia. This information can help in targeting marketing efforts and understanding the geographic reach of the hotel.

9. **Stay in Nights by Month**

  The number of nights stayed is highest in August and July, and lowest in January. This indicates peak travel seasons during the summer months and a lull during the winter months, which can influence pricing and promotional strategies.

10. **Hotel Booking by Market Segment**

  Online travel agencies (OTA) account for the highest number of hotel bookings, while aviation and undefined segments have the least. This highlights the importance of maintaining strong relationships with OTAs and potentially exploring opportunities to increase direct bookings.
